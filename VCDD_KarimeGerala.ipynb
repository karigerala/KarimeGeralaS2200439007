```python
#KARIME GERALA S2200439007

#libraries needed
import numpy as np
import matplotlib.pyplot as plt
import math as mt

#Variables
g = 9.80665
t0 = 0
tstep = 0.1
s0 = 0

#Input Variables
v0 = float(input ("Insert velocity of the car before breaking km/h:"))
print('''
Insert the next variables according to the road conditions:
concrete,dry: μ = 0.5
concrete, wet: μ = 0.35
ice, dry: μ = 0.15
ice, wet: μ = 0.08
water, aquaplaning: μ = 0.05
gravel, dry: μ = 0.035
sand, dry: μ = 0.3
''')
μ = float(input ("Insert the friction coefficient of the car:"))
m = float(input ("Insert the mass of the vehicle in kg:"))
theta = float(input("Instert the angle of inclination of the road in degrees:"))

#Creation of arrays
t_array = [t0]
s_array = [s0]           
v_array = [v0]

#conversion of units
v0= (v0 *1000)/3600   

#Constant variables
Fi = m*g*mt.sin(theta) #force of inclination
Fr = μ*((m*g)/(mt.cos(theta))) #friction force
Ft = Fi - Fr #total force 
a = Ft/m #acceleration

#Derived equations for breaking distance
sf = (1/2)*((v0**2)/(g*μ))#final breaking distance

#loop for obtainingthe simulation of the distance and velocity of the car after breaking
for i in range (1000):
    t = t0 + tstep
    v = v0 + (a*tstep)
    s = s0 + (v*tstep) + ((1/2)*a*(tstep**2))
   
    if v>=0:
        t0 = t
        v0 = v
        s0 = s
    else: 
        break
    
    t_array.append(t)
    s_array.append(s)
    v_array.append(v)

#Plot
plt.subplot(211)
plt.plot(t_array,s_array)
plt.title("Vehicle behavior in a breaking situation")
plt.ylabel("distance(m)")
plt.subplot(212)
plt.plot(t_array,v_array)
plt.xlabel("time(s)")
plt.ylabel("velocity(m/s)")
```
